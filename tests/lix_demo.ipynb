# LIX Readability — Norwegian Text Difficulty Bands

Interactive exploration of LIX/RIX scores across the difficulty spectrum.

| LIX Score | Difficulty |
|-----------|------------|
| < 25      | Very Easy  |
| 25–34     | Easy       |
| 35–44     | Medium     |
| 45–54     | Hard       |
| > 55      | Very Hard  |

```python
from lix import compute
```

## Very Easy (LIX < 25)
Short sentences, simple words. Children's book level.

```python
text = "Jeg er her. Du er der. Vi er to."
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

## Easy (LIX 25–34)
Everyday language. Newspaper tabloid level.

```python
text = (
    "Han jobbet lenge med den store oppgaven "
    "og fikk til slutt et godt og viktig "
    "resultat til dem alle."
)
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

## Medium (LIX 35–44)
Standard prose. Newspaper or popular non-fiction.

```python
text = (
    "Forskerne har oppdaget et nytt materiale "
    "som kan erstatte plast. "
    "Materialet testes på et laboratorium i dag."
)
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

## Hard (LIX 45–54)
Formal language. Government reports, official documents.

```python
text = (
    "Ministeren drøftet de nye reglene. "
    "Partene var stort sett enige om "
    "punktene i avtalen."
)
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

## Very Hard (LIX > 55)
Dense academic or legal prose. Norwegian compound words push scores high.

```python
text = (
    "Grunnlovsbestemmelsene vedrørende "
    "kommuneforvaltningens "
    "organisasjonsstruktur diskuteres "
    "inngående mellom "
    "stortingsrepresentantene og "
    "departementsrådgiverne."
)
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```
