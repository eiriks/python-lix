# LIX Readability — Norwegian Text Difficulty Bands

Interactive exploration of LIX/RIX scores across the difficulty spectrum.

| LIX Score | Difficulty |
|-----------|------------|
| < 25      | Very Easy  |
| 25–34     | Easy       |
| 35–44     | Medium     |
| 45–54     | Hard       |
| > 55      | Very Hard  |

```python
from lix import compute
```

## Very Easy (LIX < 25)
Short sentences, simple words. Children's book level.

```python
text = "Jeg er her. Du er der. Vi er to."
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

## Easy (LIX 25–34)

```python
text = (
    "Han jobbet lenge med den store oppgaven "
    "og fikk til slutt et godt og viktig "
    "resultat til dem alle."
)
result = compute(text, language="nb")
print(f"LIX:  {result.lix:.1f}")
print(f"RIX:  {result.rix:.1f}")
print(f"Band: {result.difficulty.value}")
print(
    f"Words: {result.word_count}, "
    f"Sentences: {result.sentence_count}, "
    f"Long words: {result.long_word_count}"
)
```

# GEO — Generative Engine Optimization Analysis

Measures how well your text is optimized for visibility in AI systems
(ChatGPT, Perplexity, Google AI Overviews, Claude).

Based on [Aggarwal et al. (2024)](https://arxiv.org/abs/2311.09735) — the first
academic study measuring what makes content get cited by generative engines.

| Signal | What it measures | AI visibility impact |
|--------|-----------------|---------------------|
| `statistics_density` | Data points per sentence (%, $, millions) | **+37%** |
| `citation_density` | Source attributions per sentence | **+30%** |
| `definition_density` | Definitional patterns per sentence | **+28%** |
| `quotation_density` | Quoted passages per sentence | **+28%** |
| `authority_score` | Fraction of authoritative vocabulary | **+24.5%** |
| `question_ratio` | Fraction of sentences that are questions | Extractability |
| `list_density` | List/bullet items per sentence | Extractability |
| `freshness_score` | Temporal references per sentence | Recency |

```python
from lix import analyze_geo
```

## Example 1: Plain conversational text (low GEO)
Simple text with no data, no sources, no structure. AI systems have little reason to cite this.

```python
plain = "The cat sat on the mat. It was warm and soft. The sun was shining outside."

result = analyze_geo(plain)
print("=== Plain text (low GEO) ===")
print(f"Statistics density:  {result.statistics_density:.2f}")
print(f"Citation density:    {result.citation_density:.2f}")
print(f"Definition density:  {result.definition_density:.2f}")
print(f"Quotation density:   {result.quotation_density:.2f}")
print(f"Authority score:     {result.authority_score:.2f}")
print(f"Question ratio:      {result.question_ratio:.2f}")
print(f"List density:        {result.list_density:.2f}")
print(f"Freshness score:     {result.freshness_score:.2f}")
print(f"({result.word_count} words, {result.sentence_count} sentences)")
```

## Example 2: Data-rich article (high GEO)
Text packed with statistics, citations, definitions, and quotes — exactly what AI systems love to cite.

```python
rich = (
    "According to a 2024 study by Aggarwal et al., content optimized for "
    "generative engines saw up to 41% higher visibility in AI-generated responses. "
    'The lead researcher stated "these findings represent a fundamental shift '
    'in how we think about content strategy" at the ACM SIGKDD conference. '
    "GEO is defined as the practice of optimizing web content so that it is "
    "more likely to be surfaced by AI-powered search systems. "
    "The research evaluated 9 optimization strategies across $10,000 worth of experiments."
)

result = analyze_geo(rich)
print("=== Data-rich article (high GEO) ===")
print(f"Statistics density:  {result.statistics_density:.2f}  (target: > 0.3)")
print(f"Citation density:    {result.citation_density:.2f}  (target: > 0.3)")
print(f"Definition density:  {result.definition_density:.2f}  (target: > 0)")
print(f"Quotation density:   {result.quotation_density:.2f}  (target: > 0)")
print(f"Authority score:     {result.authority_score:.2f}  (target: > 0.03)")
print(f"Question ratio:      {result.question_ratio:.2f}")
print(f"List density:        {result.list_density:.2f}")
print(f"Freshness score:     {result.freshness_score:.2f}  (target: > 0)")
print(f"({result.word_count} words, {result.sentence_count} sentences)")
```

## Example 3: Q&A format with lists
Structured content that AI systems can easily extract and quote.

```python
qa = (
    "What is the LIX readability score? "
    "LIX is defined as a readability formula developed by Bj\u00f6rnsson in 1968. "
    "It measures text difficulty using two components.\n"
    "- Average sentence length in words.\n"
    "- Percentage of long words with more than six characters.\n"
    "Why does readability matter? "
    "Research shows that 73% of readers abandon text they find too difficult."
)

result = analyze_geo(qa)
print("=== Q&A with lists ===")
print(f"Statistics density:  {result.statistics_density:.2f}")
print(f"Citation density:    {result.citation_density:.2f}")
print(f"Definition density:  {result.definition_density:.2f}")
print(f"Question ratio:      {result.question_ratio:.2f}  <- Q&A structure!")
print(f"List density:        {result.list_density:.2f}  <- structured content!")
print(f"Freshness score:     {result.freshness_score:.2f}")
print(f"({result.word_count} words, {result.sentence_count} sentences)")
```

## Example 4: Side-by-side comparison
Compare two versions of the same content — one plain, one GEO-optimized.

```python
before = (
    "Remote work is popular now. "
    "Many companies allow it. "
    "Workers seem to like it a lot."
)

after = (
    "According to a 2024 Stanford study, 73% of companies now offer "
    "remote work options. Remote work is defined as the practice of "
    "performing job duties outside a traditional office environment. "
    "Research shows that employee satisfaction increased by 2x "
    "in organizations that adopted flexible policies."
)

r_before = analyze_geo(before)
r_after = analyze_geo(after)

print(f"{'Signal':<22} {'Before':>8} {'After':>8}  {'Change':>8}")
print("-" * 50)
for field in [
    "statistics_density", "citation_density", "definition_density",
    "quotation_density", "authority_score", "question_ratio",
    "list_density", "freshness_score",
]:
    b = getattr(r_before, field)
    a = getattr(r_after, field)
    diff = a - b
    arrow = "\u2191" if diff > 0 else ("\u2193" if diff < 0 else " ")
    print(f"{field:<22} {b:>8.2f} {a:>8.2f}  {arrow} {diff:+.2f}")
```

## Example 5: Norwegian text
GEO patterns (statistics, citations) work across languages — the regex patterns detect numbers and common attribution phrases.

```python
norsk = (
    "If\u00f8lge en rapport fra SSB (2024) har 68% av norske bedrifter "
    "tatt i bruk kunstig intelligens i sine arbeidsprosesser. "
    "Forskere ved NTNU har p\u00e5vist at produktiviteten \u00f8kte "
    "med 3x i avdelinger som brukte AI-verkt\u00f8y."
)

result = analyze_geo(norsk, language="nb")
print("=== Norwegian GEO analysis ===")
print(f"Statistics density:  {result.statistics_density:.2f}  (68%, 3x)")
print(f"Citation density:    {result.citation_density:.2f}")
print(f"Freshness score:     {result.freshness_score:.2f}  (2024)")
print(f"Authority score:     {result.authority_score:.2f}")
print(f"({result.word_count} words, {result.sentence_count} sentences)")
```
